```python
import tensorflow as tf
scalar=tf.constant(5)
print(f"scalar:{scalar}")
vector=tf.constant([10,10])
print(f"vector:{vector}")
matrix=tf.constant([[10,7],[7,10]])
print(f"matrix:{matrix}")
tf1=tf.random.uniform([5,300],minval=0,maxval=1)
tf2=tf.random.uniform([5,300],minval=0,maxval=1)
print(f"tf:{tf1}")
```

```python
print(f"scalar shape:{scalar.shape}")
print(f"scalar rank:{tf.rank(scalar)}")
print(f"scalar size:{tf.size(scalar)}\n")

print(f"vector shape:{vector.shape}")
print(f"vector rank:{tf.rank(vector)}")
print(f"vector size:{tf.size(vector)}\n")

print(f"matrix shape:{matrix.shape}")
print(f"matrix rank:{tf.rank(matrix)}")
print(f"matrix size:{tf.size(matrix)}\n")

print(f"tf shape:{tf1.shape}")
print(f"tf rank:{tf.rank(tf1)}")
print(f"tf size:{tf.size(tf1)}\n")

print(f"tf shape:{tf.shape(tf2)}")
print(f"tf rank:{tf.rank(tf2)}")
print(f"tf size:{tf.size(tf2)}\n")
```

```python
tf_mal=tf.matmul(tf1,tf2,transpose_b=True)
print(f"tf_mal:{tf_mal}\n")
tf_dot = tf.reduce_sum(tf1*tf2)
print(f"tf_dot:{tf_dot}\n")
tf_dot2=tf.tensordot(tf1,tf2,axes=[[1],[1]])#different way each row of tf1 with each row of tf2
print(f"tf_dot2:{tf_dot2}\n")
```

```python
tf3=tf.random.uniform([224,224,3],minval=0,maxval=1)
tf_max_axis1=tf.reduce_max(tf3,axis=0)#along first axis
tf_min_axis1=tf.reduce_min(tf3,axis=0)
tf_max=tf.reduce_max(tf_max)
tf_min=tf.reduce_min(tf_min)
print(f"tf_max:{tf_max}")
print(f"tf_min:{tf_min}\n")
print(f"tf_max_axis1:{tf_max_axis1}")
print(f"tf_min_axis1:{tf_min_axis1}\n")
```

```python
tf4=tf.random.uniform([1,224,224,3],minval=0,maxval=1)
tf_squeezed=tf.squeeze(tf4)
print(f"tf_squeezed:{tf_squeezed.shape}")
```

```python
tf5=tf.range(1,11)
print(f"tf5:{tf5}")
tf5_max_index=tf.argmax(tf5)
print(f"tf5 max index:{tf5_max_index}")
tf_one_hot=tf.one_hot(tf5,depth=9)
print(f"tf_one_hot:{tf_one_hot}")
```
